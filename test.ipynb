```python
import matplotlib.pyplot as plt
import numpy as np

# Create a sample figure to access the colormap
fig, ax = plt.subplots(figsize=(4, 1))
cmap = plt.get_cmap('cividis')

# Create a colormap index from 0 to 255 (256 discrete colors)
color_index = np.linspace(0, 255, 256, dtype=int)

# Get the RGB values of the colormap at each index
cividis_rgb_values = cmap(color_index)

# Close the sample figure
plt.close(fig)
```

```python
import copy
def foo(original_list):
    copylist= copy.copy(original_list)  # 여기서 내가 새로운 리스트를 만들었다. 하지만 내용물의 reference는 original과 같아야하지만 여기서 내용물은 그냥 정수에 불과하다. 작은 정수는 모두 같은 memory를 가리킨다*
    copylist[0]=100

original_list = [1,2,3]

foo(original_list)
print(original_list)
```

```python
original_list = [[1, 2], [3, 4]]

# Create a shallow copy
shallow_copy = copy.copy(original_list)

# Modify an element within the shallow copy
shallow_copy[0][0] = 99

# Check the original list
print(original_list)  # Output: [[99, 2], [3, 4]]
```
